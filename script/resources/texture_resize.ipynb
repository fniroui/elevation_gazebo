```python
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt
from PIL import Image

img = Image.open('texture.jpg')
x, y = img.size
if x>y:
    img = img.crop((0,0,y,y)) 
else:
    img = img.crop((0,0,x,x)) 
img.resize((1024,1024))
print(img.size)
plt.imshow(img)
plt.show()
img.save("../worlds/models/textures/rubble.png")
```
